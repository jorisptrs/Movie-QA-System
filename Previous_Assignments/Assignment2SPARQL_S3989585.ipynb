Test the function on your 10 queries to see whether they all give the expected answers.

```python
import requests

def yourfunction(query) : 
    url = 'https://query.wikidata.org/sparql'
    results = requests.get(url, params={'query': query, 'format': 'json'}).json()
    try:
        print(results['boolean'])
    except:
        for item in results['results']['bindings']:
            for var in item :
                print('{}\t{}'.format(var,item[var]['value']))
```
